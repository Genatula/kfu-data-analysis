Создадим трёхмерный массив размерности (6, 7, 8), заполненный случайными числами. Найдём позицию (x, y, z) сотого элемента:

Создадим вектор размерности 100 со значениями от 0 до 10. Если значение находится между 3 и 8, то поменяем знак у элемента

```python
vector_from_0_to_10 = np.random.default_rng(10).random(100) * 10
print(vector_from_0_to_10)
with np.nditer(vector_from_0_to_10, op_flags=['readwrite']) as it:
    for x in it:
        if 3 <= x <= 8:
            x[...] = (-1) * x
print(vector_from_0_to_10)
```

Создадим матрицу 5x5 со значениями от 0 до 4

```python
matrix_5_x_5 = np.zeros((5, 5))
counter = 0
for row in matrix_5_x_5:
    row += counter
    counter += 1
print(matrix_5_x_5)
```

Создадим матрицу 10x10 со случайными значениями, вычислим среднее и вычтем из каждой строки:

```python
matrix_10_x_10 = np.random.default_rng(10).random((10, 10))
print(matrix_10_x_10)
mean = np.mean(matrix_10_x_10)
print(mean)
for row in matrix_10_x_10:
    row -= mean
print(matrix_10_x_10)
```
